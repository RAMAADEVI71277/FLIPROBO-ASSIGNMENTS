# Splitting the data as X and Y 

```python
x=dfnew.iloc[:,:-1]
x
```

```python
y=dfnew['Which of the Indian online retailer would you recommend to a friend?']
y
```

# Balancing the input variable (X) and output variable(Y)

```python
import sklearn
from sklearn.preprocessing import power_transform
```

```python
newx=power_transform(x)
newx
```

```python
dfx=pd.DataFrame(newx)
dfx
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smt=SMOTE()
x_train1,y_train1=smt.fit_resample(dfx,y)
```

```python
y_train1.value_counts()
```

```python
sns.countplot(y_train1)
```

# TRAINING AND TESTING THE DATA

```python
from sklearn.model_selection import train_test_split
```

# MODEL SELECTION

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
lr=LinearRegression()
```

```python
maxR2S=0
maxR=0
for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x_train1,y_train1,test_size=0.20,random_state=i)
    lr.fit(x_train,y_train)
    predlr=lr.predict(x_test)
    score=r2_score(y_test,predlr)
    print('R2_score' ,score, 'at random_state ' ,i)
    if score > maxR2S:
        maxR2S=score
        maxR=i
        print('Max r2_score ' ,maxR2S, 'at random_state ' ,i)       
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_train1,y_train1,test_size=0.20) #at default random_size as all values is the same at all random_state
lr.fit(x_train,y_train)
```

# PREDICTION

```python
predlr=lr.predict(x_test)
score=r2_score(y_test,predlr)
print('Best r2_score ' ,score*100,'of model ' ,lr)
```

# SAVING THE BEST MODEL

```python
import pickle
filename='customer_retention_dataset.pkl'
pickle.dump(lr,open(filename,'wb'))
```
